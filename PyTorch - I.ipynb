neural network oluşturmak için gerekli diğer yapıyı ekleyelim

```python
import torch.nn as nn
```

tensor oluşturalım

```python
input_tensors = torch.tensor(
    [[0.3471,0.4547,-0.2356]]
)
```

doğrusal bir katman olan Linear Layer ekleyelim

```python
linear_layer = nn.Linear(
    in_features = 3,
    out_features= 2
)
```

output görelim

```python
output = linear_layer(input_tensors)
print(output)
```
